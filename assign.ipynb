```python
n = int(input("enter how many numbers: "))

numbers = []

for i in range(n):
    x = int(input("enter a number: "))
    numbers.append(x)

maximum = numbers[0]
minimum = numbers[0]

for i in range(n):
    if numbers[i] > maximum:
        maximum = numbers[i]
    if numbers[i] < minimum:
        minimum = numbers[i]

print("maximum =", maximum)
print("minimum =", minimum)
```

```python
s = input("Enter a string: ")

left = 0
right = len(s) - 1

is_palindrome = True

while left < right:
    if s[left] != s[right]:
        is_palindrome = False
        break
    left = left + 1
    right = right - 1

if is_palindrome:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

```python
n = int(input("Enter how many numbers: "))
arr = []

for i in range(n):
    a = int(input("Enter a number: "))
    arr.append(a)

x = int(input("Enter the value of x: "))

arr.sort()

left = 0
right = 1
found = False

while right < n:
    diff = arr[right] - arr[left]

    if diff == x:
        print("Pair found:", arr[left], "and", arr[right])
        found = True
        break

    
    if diff < x:
        right = right + 1
    else:
        left = left + 1

    
    if left == right:
        right = right + 1

if not found:
    print("No pair found")   
    
```
