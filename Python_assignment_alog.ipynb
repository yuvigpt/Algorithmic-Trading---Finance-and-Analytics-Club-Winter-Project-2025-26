```python
st="gunnug"
left=0
right=len(st)-1
while(left<right):
    if(st[left]!=st[right]):
      print("Not Palindrome")
      break
    left+=1
    right-=1
else:
  print("Palindrome")
```

```python
arr=np.array([1,0,2,2,0,0,1,1])
low=0
mid=0
high=len(arr)-1

while(mid<=high):
    if(arr[mid])==0:
      arr[mid],arr[low]=arr[low],arr[mid]
      mid+=1
      low+=1
    elif(arr[mid]==1):
      mid+=1
    else:
      arr[mid],arr[high]=arr[high],arr[mid]
      high-=1
print(arr)
```
