```python
# Question 1

arr = [int(x) for x in input("Enter numbers: ").split()]

maximum = arr[0]
minimum = arr[0]

for i in range(1, len(arr)):
  if arr[i] > maximum:
    maximum = arr[i]

  if arr[i] < minimum:
    minimum = arr[i]


print("Maximum: ", maximum)
print("Minimum: ", minimum)


```

```python
# Question 2

a = input("Enter a word: ")

left = 0
right = len(a) - 1

is_palindrome = True

while left<right:
  if a[left] != a[right]:
    is_palindrome = False
    break

  left += 1
  right -= 1
if is_palindrome:
  print("The given word is a palindrome")

else:
  print("The given word is not a palindrome")
```
