```python
#Question 1
arr = [int(x) for x in input("Enter numbers: ").split()]

max = arr[0]
min = arr[0]

for i in range(1, len(arr)):
  if arr[i] > max:
    max = arr[i]
  if arr[i] < min:
    min = arr[i]

print("Maximum =",max)
print("Minimum =", min)
```

```python
#Question 2
a = input("Enter a string: ")

left = 0
right = len(a) - 1

is_palindrome = True

while left < right:
  if a[left] != a[right]:
    is_palindrome = False
    break
  left += 1
  right -= 1

if is_palindrome:
  print("The given string is a palindrome.")
else:
  print("The given string is not a palindrome")
```
