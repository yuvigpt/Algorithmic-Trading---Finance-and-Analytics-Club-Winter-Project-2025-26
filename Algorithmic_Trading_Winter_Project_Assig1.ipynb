 The first question involves arithmetic with a small application of loops. You are given an array of n integers, that is, a1,a2,...,an, and you are require to output the maximum and minimum among them.
 Hint: Try with n = 3, then generalise your argument

```python
import numpy as np

array1 = np.array([10,20,30])
max_no = array1[0]
length = len(array1)
for x in range(0,length):
  max_no = max(max_no,array1[x])
print(max_no)
```

 This question is for familiarity with strings. You are needed to input a string and check if it is a palindrome.
 Hint: Read a bit about the two-pointer method.

```python
string = input("Write something ")
length = len(string)
flag = True
for x in range(0,length//2):
  if string[x] != string[length - 1 - x]:
    flag = False
    print("It is not a palindrome")
    break
  else:
    continue
if flag == True:
  print("It is a palindrome")
```

 You will need a basic understanding of sorting for this question. Note that there is a direct Python function to sort a sequence of integers. You are given an array of n integers, that is,a1, a2,...,an, and a number x. You need to find a pair of elements whose absolute difference is x. You are assured that such a pair exists in the array. Hint: What if you sort the array?

```python
array3 = [int(x) for x in input("Enter the nos. & press 'ENTER' to end ").split()]
x = int(input("Enter the difference to find: "))
array3.sort()
p1 = 0
p2 = 1
flag = True
while p2 < len(array3):
    diff = array3[p2] - array3[p1]
    if diff == x:
        print("Pair found:", array3[p1], array3[p2])
        flag = False
        break
    elif diff < x:
        p2 += 1
    else:
        p1 += 1
    if p1 == p2:
        p2 += 1
if flag == True:
  print("No such pair exist")
```

You are given an array of length n, and distinct integers a, b, and c. All elements of the array are one of a, b, or c. You are required to sort the array without the use of a conventional sorting algorithm. Hint: Think of three pointers.

```python
array3 = [x for x in input("Enter 'a', 'b' or 'c' & press 'ENTER' to end ").split()]
ap = 0
bp = 0
cp = len(array3) - 1
while bp <= cp:
    if array3[bp] == 'a':
      array3[ap], array3[bp] = array3[bp], array3[ap]
      ap += 1
      bp += 1
    elif array3[bp] == 'b':
      bp += 1
    else:
      array3[bp], array3[cp] = array3[cp], array3[bp]
      cp -= 1
print(array3)
```
