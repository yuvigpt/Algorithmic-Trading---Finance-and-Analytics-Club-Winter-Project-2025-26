# using loc and iloc function to access the row "c" and print of given table



```python
data1 = {'one':pd.Series([1,2,3], index = ['a', 'b', 'c']),
        'two':pd.Series([1,2,3,4], index = ['a', 'b', 'c', 'd'])}
table= pd.DataFrame(data1)
print(table)
```

#The append() function can be used to add more rows to the DataFrame. so add 3 more rows on table and print table

```python
a = {
    'one': pd.Series([5,6,7], index=['e','f','g']),
    'two': pd.Series([8,9,3], index=['e','f','g'])
}
new_rows = pd.DataFrame(a)
table = pd.concat([table, new_rows])
print(table)
```

#The drop() function can be used to drop rows whose labels are provided so drop 3rd row

```python
table=table.drop('c',axis=0)
print(table)
```

# plot the function y = xsqaure , x3 and x raised to power 4 in same plot in the range of x=10;      for example image.png

```python
fig2=plt.figure()
x = np.arange(0,10)
y = x**2
z=x**3
a=x**4
axes=fig2.add_axes([0,0,1,1])
plt.xlim(0, 4.0)
plt.ylim(0,260)
axes.plot(x,y,color='blue')
axes.plot(x,z,color='red')
axes.plot(x,a,color='green')
plt.show()
```

# implement customization options for markers on above graph                     
 like thisimage.png

```python
fig2=plt.figure()
x = np.arange(0,10)
y = x/8+1
z=x/8+1.5
a=x/8+2
b=x/8+3
c=x/8+4
axes = fig2.add_axes([0,0,1,1])
axes.set_xlim(0,10)
axes.set_ylim(0,6)
axes.plot(x,y,'b*')
axes.plot(x, z, marker='o', color='orange', linestyle='None')
axes.plot(x,a,'gD')
axes.plot(x,b,'r^')
axes.plot(x, c, marker='s', color='purple', linestyle='None')
plt.show()
```

# plot a Bar Chart of volume of trades for last 15 days of your favourite stock(extract data from google or yfinance)
for exampleimage.png

```python
from datetime import datetime
import seaborn as sns
import yfinance as yf
import matplotlib.pyplot as plt
ticker = "NVDA"
end_date = datetime(year=2025, month=12, day=7)
start_date = datetime(year=2025, month=11, day=20)
df_single = yf.download(
    tickers=ticker,
    start=start_date,
    end=end_date,
    interval="1d",
    group_by="ticker",
    auto_adjust=True,
    progress=False
)
print(df_single)
import matplotlib.pyplot as plt
import pandas as pd

df = df_single['NVDA'].reset_index()
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
plt.figure(figsize=(12,6))
plt.bar(df['Date'], df['Volume'], width=0.6)
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("NVDA Daily Trading Volume (Trading Days Only)")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

#plot a pie chart of market share of stockbrokers.(take data from google)

image.png

```python
import matplotlib.pyplot as plt

labels = [
    "Groww", "Zerodha", "Angel", "RSKV (Upstox)",
    "ICICI Securities", "Kotak Securities",
    "HDFC Securities", "Motilal Oswal", "Others"
]
shares = [25.8, 16.5, 15.4, 5.8, 4, 3, 2.7, 2, 24.8]
plt.figure(figsize=(9,9))
wedges, texts, autotexts = plt.pie(
    shares,
    labels=labels,
    autopct="%.1f%%",
    startangle=90,
    pctdistance=0.8
)
circle = plt.Circle((0,0), 0.55, color='white')
plt.gca().add_artist(circle)
plt.title("Market Share")
plt.tight_layout()
plt.show()

```
