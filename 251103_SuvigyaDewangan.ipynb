```python
def get_min_max(a):
    max = min = a[0]
    for i in range(0, len(a)):
        if a[i] > max:
            max = a[i]
        elif a[i] < min:
            min = a[i]
        else:
            continue
    print(f'Maximum is {max} and minimum is {min}.')


#def get_max_min(a):
#    print(f'Maximum is {max(a)} and minimum is {min(a)}.')


n = int(input("Enter number of values to be added: "))
arr = []
for i in range (0, n) :
  num = int(input("Enter a number: "))
  arr.append(num)

get_min_max(arr)
#get_max_min(arr)
```

```python
def is_palindrome_2pointer(s):
    i = 0
    j = len(s) - 1
    is_valid = 1
    while i <= j:
        if s[i] != s[j]:
            is_valid = 0
        i += 1
        j -= 1

    return is_valid == 1


#def is_palindrome(s):
#    return s == s[::-1]


string = input("Enter a string: ")

#print(is_palindrome(string))
print(is_palindrome_2pointer(string))
```

Solution 3

```python
def find_pair(a, x):
    i = 0
    j = 1
    a.sort()
    while i < len(a) and j < len(a):
        if a[j] - a[i] > x:
            i += 1
        elif a[j] - a[i] < x:
            j += 1
        else:
            return [a[i], a[j]]

arr = []
n = int(input("Enter the number of values to be added: "))
arr = []
for i in range (0, n) :
  num = int(input("Enter a number: "))
  arr.append(num)
x = int(input("Enter the value of required difference: "))

print(find_pair(arr, x))
```

Solution 4

```python
def sort_arr(arr, a, b, c):
    low = 0
    mid = 0
    high = len(arr) - 1

    while mid <= high:
        if arr[mid] == a:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == b:
            mid += 1
        elif arr[mid] == c:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

    return arr


n = int(input("Enter the number of values to be added: "))
a = int(input("Enter the first value: "))
b = int(input("Enter the second value: "))
c = int(input("Enter the third value: "))
arr = []
for i in range (0, n) :
  num = int(input("Enter the number: "))
  arr.append(num)

print(sort_arr(arr, a, b, c))
```
