```python
def findpair(numbers, x):
    numbers.sort()
    n = len(numbers)

    for i in range(n):
        for j in range(i + 1, n):
            diff = numbers[j] - numbers[i]

            if diff == x:
                return numbers[i], numbers[j]
            elif diff > x:
                break

arr = [5, 20, 3, 2, 50, 80]
diff = 78

result = findpair(arr,diff)
print(result)
```
